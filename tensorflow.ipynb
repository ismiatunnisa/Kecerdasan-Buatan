tensor dimension

```python
# dimension display
arr_ran.shape
```

```python
#reshaping
print(arr_ran.shape)
arr_ran1 = tf.reshape(arr_ran, [2,2,221184])
print(arr_ran1.shape)
```

```python
#expension(memperluas)
print(arr_ran.shape)
arr_ran1 = tf.expand_dims(arr_ran,axis=5)
print(arr_ran1.shape)
```

```python
#squeezing(memperkecil)
print(arr_ran1.shape)
arr_ran2 = tf.squeeze(arr_ran1, axis=5)
print(arr_ran2.shape)
```

```python
#transpose(baris jadi kolom)
print(arr_ran.shape)
arr_ran1 = tf.transpose(arr_ran)
print(arr_ran1.shape)
```

```python
#broadcast(array diulang2)
print(arr_ran.shape)
arr_ran1 = tf.broadcast_to(arr_ran, [12,6, 6, 12, 32, 64])
print(arr_ran1.shape)
```

tensor concatenantion and spliting

```python
#conca
#concat(bergabung)
arr_random1 = tf.random.normal((3,3,3), seed=1)
arr_random2 = tf.random.normal((3,3,3), seed=5)
```

```python
tf.concat([arr_random1, arr_random2], axis=0)
```

```python
arr_random2
```

```python
#stack(dipisah)
tf.stack([arr_random1, arr_random2], axis=0)
```

```python
#split
arr_random_1_1= tf.concat([arr_random1, arr_random2], axis=0)
```

```python
arr_random_1_2= tf.stack([arr_random1, arr_random2], axis=0)
```

```python
temp= tf.split(arr_random_1_1, num_or_size_splits=4)
temp
```

```python
temp, temp1, temp2, temp4= tf.split(arr_random_1_1, num_or_size_splits=4)
temp2
```

```python
temp=tf.unstack(arr_random_1_1, axis=1)
temp
```
