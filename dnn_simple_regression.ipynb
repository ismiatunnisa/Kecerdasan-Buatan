```python
#simple regression
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
X = tf.constant([1,2,3,4,7,11,13,19,23,29])
Y = tf.constant([5,7,8,10,12,16,18,24,28,31])
```

```python
plt.scatter(X,Y)
plt.show()
```

### taHap modeling
1.arsitektur
2. configurasi
3. model training

```python
#bikin model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])
```

```python
model.compile(
    optimizer='adam', 
    loss='mae'
    )
```

```python
model.fit(
    x = X, y = Y,
    epochs=500, #perulangan model
    steps_per_epoch=10
)
```
